```python
import pandas as pd
import numpy as np

points = pd.read_csv("season_data/ff_points.csv")
adp = pd.read_csv("adp_data/adp.csv")
history = pd.read_csv("season_data/draft_history.csv")
```

```python
points = points[points["Year"]!=2013].reset_index(drop=True)
points
```

```python
adp.head()
```

```python
history.head()
```

```python
adp_and_draft = pd.merge(adp, history, how="inner", on="Player")
adp_and_draft["Player"] = adp_and_draft["Player"].str.replace(" ", "")
adp_and_draft
```

```python
ff_data = pd.merge(points, adp_and_draft, how="inner", on=["Player", "Year"])
ff_data.head()
```

```python
ff_data["Experience"] = ff_data["Year"] - ff_data["draft_year"]
ff_data.head()
```

```python
ff_data["Rookie"] = (ff_data["Experience"]==0)*1
ff_data["Soph"] = (ff_data["Experience"]==1)*1
ff_data["Third"] = (ff_data["Experience"]==2)*1
```

```python
ff_data.to_csv("ff_data.csv", encoding='utf-8', index=False)
```
