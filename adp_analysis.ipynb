```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

pd.options.display.max_columns = 100
```

```python
ff_data = pd.read_csv("ff_data.csv")
```

```python
for column in ff_data.columns.tolist()[4:17]:
    ff_data[column] = np.round((ff_data[column]/ff_data["G"]), 2)
```

```python
ff_data
```

```python
sns.set(rc={"figure.figsize":(20, 12)})
sns.relplot(data=ff_data, x="adp", y="PPR", col="FantPos", kind="scatter");
```

```python
ff_data["adp_round"] = np.round(((ff_data["adp"] / 12)+1), 1)
ff_data.head()
```

```python
import math

ff_data["adp_round"] = [np.round(math.ceil(i * 3)/3, 1) for i in ff_data["adp_round"].tolist()]
```

```python
ff_adp_round = ff_data[["FantPos", "PPR", "adp_round"]]
ff_adp_round = ff_adp_round.groupby(["FantPos","adp_round"], as_index=False).agg(np.mean)
```

```python
sns.relplot(data=ff_adp_round, x="adp_round", y="PPR", col="FantPos", kind="line");
```

```python
from itertools import combinations
from scipy import stats

def ttest_run(data, pos, c1, c2):
    results = stats.ttest_ind(data[(data["FantPos"]==pos)&(data["adp_round"]==c1)]["PPR"],
                              data[(data["FantPos"]==pos)&(data["adp_round"]==c2)]["PPR"])
    df = pd.DataFrame({'categ1': c1,
                       'categ2': c2,
                       'tstat': results.statistic,
                       'pvalue': results.pvalue}, 
                       index = [0])    
    return df

df_list = [ttest_run(ff_data, "RB", i, j) for i, j in combinations(ff_data[ff_data["FantPos"]=="RB"]['adp_round'].unique().tolist(),2)]

rb_diff_in_means = pd.concat(df_list, ignore_index = True)
rb_diff_in_means
```
